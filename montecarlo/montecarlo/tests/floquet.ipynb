```julia
N = 4
T = 2
tgrid = 0:.01:N*T
drive_f(t) = cos.(t*2*pi/T)
drive_g(t) = cos.(t*2*pi/T/N) + sin.(t*2*pi/T/N*5)
# plot(tgrid, cos.(tgrid*2*pi/T))
plot(tgrid, [drive_f(tgrid), drive_g(tgrid), drive_f(tgrid)+ drive_g(tgrid)], label=["Floquet" "Gate" "Total Drive"])
# Tgrid = 0:N
# scatter!(Tgrid, drive_f(Tgrid)+drive_g(Tgrid))
```
